2) One-Hot encoding



```python
categorical_columns = df_adult_data.select_dtypes('object').drop(columns=['income']).columns.tolist() #виокремили категоріальні фічі
encoder = OneHotEncoder(sparse_output=False)                                                          #створили об'єкт
one_hot_encoded = encoder.fit_transform(df_adult_data[categorical_columns])                           #заповнили
one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(categorical_columns), index=df_adult_data.index) #перетворили на об'єкт pandas
df_encoded = pd.concat([df_adult_data, one_hot_df], axis=1)                                           #склеїли з основою по тсовпчикам
df_encoded = df_encoded.drop(categorical_columns, axis=1)                                             #дропнули минулі колонки з якісними фічами
df_encoded
```

Занадто багато колонок


```python
zero_perс = (df_encoded.drop(columns=['age','fnlwgt','education-num',	'capital-gain',	'capital-loss', 'hours-per-week','income']) == 0).mean() * 100
zero_perс.tolist()
```

Видалимо всі рядки,  в яких відстуні >90% значень

```python
columns_with_zeros = zero_perс[zero_perс > 90].index
df_encoded_without_zeros = df_encoded.drop(columns=columns_with_zeros)
df_encoded_without_zeros
```

```python
zero_all = (df_encoded_without_zeros.drop(columns=['income']) == 0).mean() * 100
zero_all
```

#13)

Маасштабування потрібне наприклад значення фіч 'age', 'fnlwgt', capital-gain	capital-loss різняться масштабом



Нормалізація —  коли для даних з відомими межами або якщо важливий діапазон.


```python
df_encoded_without_zeros
```

```python
scaler = MinMaxScaler()
df_encoded_without_zeros[['fnlwgt_normalized', 'capital-gain_normalized','capital-loss_normalized', 'age_normalized','education-num_normalized', 'hours-per-week_normalized']] = scaler.fit_transform(df_encoded[['fnlwgt', 'capital-gain', 'capital-loss', 'age', 'education-num', 'hours-per-week']])
```

```python
df_normal = df_encoded_without_zeros.drop(columns=['fnlwgt','capital-gain','capital-loss','age'	,'education-num','hours-per-week'	])
df_normal
```

Стандартизація — якщо дані мають різні масштаби та відмінності, важливо, щоб вони мали нормальне значення.

```python
scaler = StandardScaler()
df_encoded_without_zeros[['fnlwgt_standart', 'capital-gain_standart','capital-loss_standart', 'age_standart','education-num_standart', 'hours-per-week_standart']] = scaler.fit_transform(df_encoded[['fnlwgt', 'capital-gain', 'capital-loss', 'age', 'education-num', 'hours-per-week']])
```
